```python
from ivhd import IVHD
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt

dataset = load_digits()
dataset
X = dataset["data"]
y = dataset["target"]
X = X / 255.
```

```python
pca = PCA()
X_pca = pca.fit_transform(X)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, s=2)
```

```python
tsne = TSNE()
X_tsne = tsne.fit_transform(X)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, s=2)
```

```python
ivhd = IVHD(simulation_steps=200, nn=10, rn=10)
X_ivhd = ivhd.fit_transform(X)
plt.scatter(X_ivhd[:, 0], X_ivhd[:, 1], c=y, s=2)
```
